# All Figures

```python
from svgdatagrabber import SvgPathParser


parser = SvgPathParser(
    "praai2013.svg",
    min_segments=6,
    tolerance=1,
)

parser.parse().plot(fig_kwargs=dict(figsize=(20, 12)))
```

# Shear Stress

```python
from svgdatagrabber import SvgPathParser

parser = SvgPathParser(
    "praai2013.svg",
    xrange=(0, 3000),
    yrange=(4800, 6000),
    min_segments=6,
    tolerance=1,
)

translate = (-1221.49, -5024.66)
scale = (0.235, 0.245)
parser.parse(translate=translate, scale=scale).plot(fig_kwargs=dict(figsize=(20, 12)))
```
