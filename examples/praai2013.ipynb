# All Figures

```python
from svgdatagrabber import SvgPathParser


parser = SvgPathParser(
    "praai2013.svg",
    min_segments=6,
    tolerance=1,
)

paths = parser.parse()
paths.plot(fig_kwargs=dict(figsize=(20, 12)))
```

# Shear Stress

```python
from svgdatagrabber import SvgPathParser

parser = SvgPathParser(
    "praai2013.svg",
    xrange=(0, 3000),
    yrange=(5000, 6100),
    min_segments=6,
    tolerance=1,
)

translate = (-1221.49, -5024.66)
scale = (0.235, 0.245)
paths = parser.parse(translate=translate, scale=scale)
paths.to_csv("results/tau-sign.csv", x="sign", y="tau")
paths.plot(fig_kwargs=dict(figsize=(20, 12)))
```

# Normal Stress

```python
from svgdatagrabber import SvgPathParser

parser = SvgPathParser(
    "praai2013.svg",
    xrange=(3300, 5200),
    yrange=(4800, 6100),
    min_segments=6,
    tolerance=1,
)

translate = (-3380, -6000 + 320)
scale = (0.00715, 0.328)
paths = parser.parse(translate=translate, scale=scale, origin=(0.0, 320.0))
paths.to_csv("results/w-sign.csv", x="w", y="sign")
paths.plot(fig_kwargs=dict(figsize=(20, 12)))
```
