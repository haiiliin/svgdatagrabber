# Original Figures

```python
from svgdatagrabber import SvgPathParser


parser = SvgPathParser(
    "praai2013.svg",
    min_segments=6,
    drop_horizontal_lines=False,
    drop_vertical_lines=False,
    tolerance=1,
)

paths = parser.parse()
ax = paths.plot(fig_kwargs=dict(figsize=(20, 12)))
ax.get_legend().remove()
ax.get_figure().savefig('results/original-figure.pdf', bbox_inches='tight', pad_inches=0.1)
```

# Shear Stress

```python
from svgdatagrabber import SvgPathParser

parser = SvgPathParser(
    "praai2013.svg",
    xrange=(0, 3000),
    yrange=(5000, 6100),
    min_segments=6,
    tolerance=1,
)
parser.csys.setup_xaxis(start=(1223.5, 5022.2), end=(2910.2, 5022.2), xstart=0.0, xend=400.0, y=0.0)
parser.csys.setup_yaxis(start=(1223.5, 5022.2), end=(1223.5, 6249.3), ystart=0.0, yend=300.0, x=0.0)

paths = parser.parse()
paths.to_csv("results/sign-tau.csv", x="sign", y="tau")
ax = paths.plot(fig_kwargs=dict(figsize=(20, 12)))
ax.set_xlim(0, 400)
ax.set_ylim(0, 300)
ax.get_figure().savefig('results/sign-tau.pdf', bbox_inches='tight', pad_inches=0.1)
```
